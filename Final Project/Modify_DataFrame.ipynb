```python
import pandas as pd 
df_Price = pd.read_csv('gold_price_dayli_oanda.csv')
cols = df_Price.columns.tolist()
cols = ['Date','Close','Open','High','Low','Volume','detail','dt_percent']
cols
df_Price = df_Price[cols]
cols = ['Date','Price','Open','High','Low','Vol.','detail','dt_percent']
df_Price.columns= cols
df_Price.to_csv('gold_price_dayli_oanda_Modified.csv')
df_Price
```

```python
df_Price['Vol.']  =df_Price['Vol.']/1000
df_Price['Vol.']
```

```python
# add K end of Vol Column 
count = 0
for  num in df_Price['Vol.'] : 
       df_Price['Vol.'][count] =  (str(num)[0:-1]) + 'K'
       count += 1 
df_Price
```

```python
df_Price = df_Price.rename({'dt_percent':'Change %'},axis=1)
df_Price
```

```python
count = 0 
for num in df_Price['Change %'] : 
    try : 
        num = float(float(num) / 100 )
        num = str(num) + '%'
        df_Price['Change %'][count] = num 

    except : 
        print(num) 
        print(count) 
        print(type(num))
    count += 1 
df_Price
```

```python

df_Price.to_csv('gold_price_daily_oanda_ModifiedColumnsM.csv',index=False)
```
